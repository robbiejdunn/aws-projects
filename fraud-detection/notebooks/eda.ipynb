# Fraud detection EDA

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
# Read data into pandas Dataframe
data_location = "../data/creditcard.csv"
credit_card = pd.read_csv(data_location)
credit_card.head(5)
```

```python
credit_card.info()
```

```python
# Target variable analysis
print(credit_card["Class"].value_counts())
print()
print(credit_card["Class"].value_counts() / len(credit_card))
# sns.countplot(credit_card["Class"])
print()
print(credit_card["Class"].value_counts()[0] / credit_card["Class"].value_counts()[1])
```

```python
sns.countplot(credit_card, x="Class")
```

```python
sns.displot(credit_card, x=f"V1", bins=20, hue="Class", multiple="dodge")
plt.yscale("log")
```

```python
sns.displot(credit_card, x=f"V1", bins=20, col="Class")
plt.yscale("log")
```

```python
sns.displot(credit_card, x=f"V1", bins=20, hue="Class", stat="density", common_norm=False, multiple="dodge")
plt.yscale("log")
```

```python
for i in range(1, 29):
    sns.displot(credit_card, x=f"V{i}", bins=20, hue="Class", multiple="dodge")
    plt.yscale("log")
```
